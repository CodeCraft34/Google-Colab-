##Series
It's is one dimensional array holding data of any types.

```python
a = ['Banana','Cat','Apple']
myseries = pd.Series(a)
myseries
```

##Labels
Labels are used to specific or name any elements in a _Series_.
Also index are used too specifiy the element.

```python
#By index
myseries[1]
```

```python
myseries = pd.Series(a, index= ['x','y','z'])
#return the value of y
myseries['y']
```

```python
myseries['z']
```

##Key/Value Object in Series
You can store multiple values in a key like a __dictionary__ format.

```python
import pandas as pd

calories = {"day1": 420, "day2": 380, "day3": 390}

myvar = pd.Series(calories)

myvar
```

##Selecting Key and Values from Series
You can select value by `index` argument i.e *day1* and *day2*.

```python
myvar = pd.Series(calories, index = ["day1", "day2"])
myvar
```

##__Dataframe__
__Dataframe__ is two dimensional tabular dataset holding *Columns* and *Rows*.

```python
data = {
    'Name': ['Alice', 'Bob', 'Charlie', 'David'],
    'Age': [25, 30, 35, 40],
    'City': ['New York', 'Los Angeles', 'Chicago', 'Houston']
}
df = pd.DataFrame(data)
df
```

##Locating Rows
Now you can see the output is in tabular with columns and rows.
You can access rows by `.loc` attribute.
It will return a *Series*.

```python
df.loc[1]
```

```python
df.loc[[0,1]]
#It will return dataframe
#Use list for giving indices
```

##Labeling the Indices
You can also label indices like Series.

```python
data = {'Name': ['Tom', 'Nick', 'Krish', 'Jack'],
        'Age': [20, 21, 19, 18]}

#Create DataFrame with custom named indices 'row1' through 'row4'
df = pd.DataFrame(data, index=['row1', 'row2', 'row3', 'row4'])
df
```

##Finding missing Value
You can find missing by `.isnull()` method.
It will return *True* when data is not present and *False* when data is present.

```python
# Returns a count of missing values for each column
raw_data = pd.read_csv('raw-data.csv')
print(raw_data.isnull())
```
