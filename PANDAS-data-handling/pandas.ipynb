#Introduction to Pandas<br>

##What is __Pandas__?
<strong>Pandas</strong> is a powerful Python library used for working with data sets.

It has functions for analyzing, cleaning, exploring, and manipulating data.

The name *Pandas* has a reference to both *Panel Data*, and *Python Data Analysis* and was created by Wes McKinney in 2008.

##Why use Pandas?
* Easy for data cleaning.
* Easily handle large data set.
* Comes with built-in functions.
* Integration with other python libraries.

##Series
It's is one dimensional array holding data of any types.

```python
a = ['Banana','Cat','Apple']
myseries = pd.Series(a)
myseries
```

##Labels
Labels are used to specific or name any elements in a _Series_.
Also index are used too specifiy the element.

```python
#By index
myseries[1]
```
